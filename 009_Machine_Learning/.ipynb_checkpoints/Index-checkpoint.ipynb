- c1：预处理，编码，回归
    - 预处理 | 标准化(Standardized)，范围缩放(Scaling)，归一化(Normalization)，二值化(Binarization)
    - Label_encoder | 标记编码
    - One-Hot编码
    - Linear Regression | 线性回归
    - Ordinary Least Square(OLS) | 普通最小二乘法
    - Ridge Regression | 岭回归
    - Polynomial Regression | 多项式回归
    - Stochastic Gradient Descent Regressor(SGD) | 随机梯度下降回归
    - Decision Tree | 决策树
    - AdaBoost Decision Tree(AdaBoost - Adaptive Boosting) | 自适应提升决策树，弱学习器加权组合，集成学习算法，实例：预测房价
    - Random Forest Regressor | 随机森林回归，集成学习算法，实例：评估共享单车的需求分布

- c2：分类，交叉验证，混淆矩阵，性能报告
    - Logistics Regression Classifer | 逻辑回归分类
    - Naive Bayes Classifier(NBC) | 朴素贝叶斯分类，Gaussian Naive Bayes
    - Cross Validation | 交叉验证
    - Confusion Matrix | 混肴矩阵 
    - Model Performance | classification_report, 模型性能报告
    - Random Forest Classifier | 随机森林分类，根据汽车特征评估质量
    - Validation Curve | 模型验证曲线，迭代次数vs准确率，(超参数遍历+交叉验证 = DOE矩阵计算结果)
    - Learning Curve | 学习曲线，样本数vs准确率

- c3：SVM (非)线性核函数，样本不均衡，置信度，最优化
    - Support Vector Machine(SVM) | SVC,线性分类器 params = {'kernel': 'linear'}
    - SVM | 非线性分类器 params = {'kernel': 'poly', 'degree': 3}，params = {'kernel': 'rbf'}
    - Kernel Function | 核函数
    - Radial Basis Function(RBF) | 径向基函数
    - 'class_weight':'auto' | 处理样本不均衡
    - Confidential Interval | 分类置信区间
    - Grid_Search | 超参数自动寻优
    - SVC | SVM分类,实例：预测一栋大楼实际进出的人数
    - SVR | SVM回归，实例：估算体育场比赛期间的交通流量

    - Mean Shifting | 均值漂移聚类 
    - Hierarchical Clustering | 凝聚层次聚类

    - Density-Based Spatial Clustering of Applications with Noise(DBSCAN) | 带噪声基于密度的聚类方法，自动估算聚类数量
    - Affinity Propagatioin | 近邻传播聚类算法，实例：探索股票数据的模式
    - Cluster, Covariance, Manifold, MeanShift, Estimate_bandwidth | 实例：建立客户细分模型

- c7：语音识别

- c8：时间序列模型
    - 将数据转换为时间序列格式
    - 切分时间序列数据
    - 操作时间序列数据（过滤）
    - dataframe.rolling.mean(), dataframe.rolling.corr() 滑动窗口均值、相关系数 | 从时间序列数据中提取统计数字
    - Hidden Markov Models(HMM)，hmmlearn.hmm.GaussianHMM | 针对序列数据创建隐马尔可夫模型,HMM是生成模型，可以通过不断生成数据来逼近结果
    - Conditional Random Fields(CRF)，sklearn_crfsuite | 条件随机场，判别模型，针对序列文本数据创建条件随机场，用于命名实体识别。自动问答，知识图谱等。给定一组输入序列条件下另一组输出序列的条件概率分布模型，在自然语言中也有广泛应用
    - hmmlearn.hmm.GaussianHMM | 实例：用隐马尔可夫模型分析股票市场数据

    - 利用视觉单词码本和向量量化创建特征 | 用'视觉码本'来创建图像识别标识。在训练数据集中，码本就像是个字典，用于提出关于图像的描述

- c11：深度神经网络
    - Perceptron,neurolab | 感知机
    - neurolab支持的网络类型
        - 单层感知机(single layer perceptron)
        - 多层前馈感知机(Multilayer feed forward perceptron)
        - 竞争层(Kohonen Layer)
        - 学习向量量化(Learning Vector Quantization)
        - Elman循环网络(Elman recurrent network)
        - Hopfield循环网络(Hopfield recurrent network)
        - 卷边循环网络(Hemming recurrent network)
    - single_layer_net = nl.net.newp([[x_min, x_max], [y_min, y_max]], 2) | 单层神经网络(一个输入层、一个隐藏层和一个输出层)
    - multilayer_net = nl.net.newff([[min_value, max_value]], [10, 10, 1]) | 深度神经网络(一个输入层、多个隐藏层和一个输出层组成)
    - epochs vs error | 训练误差曲线
    - nl.net.newlvq(nl.tool.minmax(data), 10, [0.25, 0.25, 0.25, 0.25]) | 向量量化器，用神经网络做向量量化
    - net = nl.net.newelm([[-2, 2]], [10, 1], [nl.trans.TanSig(), nl.trans.PureLin()]) | 为序列数据分析创建一个递归神经网络
        - P , RQ1矩阵，包含Q1个长度为R的典型输入向量
        - T , SNQ2矩阵，包含Q2个长度为SN的典型目标向量，SN决定了输出向量的长度
        - Si，N-1个隐含层包含的神经元个数，默认值为[]
        - TFi，第i层传递函数，用字符串表示。对于输出层，默认值为‘purelin’，对于隐含层，默认值为‘tansig’
        - BTF，反向传播网络的训练函数，默认值为‘traingdx’，可取值还有‘traingd’、‘traingdm’、‘traingda’
        - BLF，反向传播权值/阙值学习函数，默认值为‘learndm’，可取值还有‘’learngd’等函数
        - PF，性能函数，默认值为‘mse’，可取值还有‘msereg’
        - IPR, 输入处理函数，默认值为‘{‘fixunknowns’，‘removeconstantrows’，‘mapminmax’}’
        - OPF,输处处理函数，默认值为‘{‘removeconstantrows’，‘mapminmax’}’
    - 实例：在光学字符识别数据库中将字符可视化。识别图像中手写字符
    - 实例：用神经网络创建一个光学字符识别器
