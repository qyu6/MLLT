主成分分析PCA
- 主成分分析（PCA-Principal components analysis）是一个降维技术，常用于计算机视觉和机器学习。当需要处理很大的特征维度时，以降低训练成本。降维时希望尽可能保留数据中的重要信息。
- PCA识别数据中的重要成分，并按照重要程度进行排序。本例展示PCA在人脸识别系统中的应用

```python
%run pca.py
```

核主成分分析（kpca - Kernel PCA）
- PCA能够实现降维，但PCA是以线性方式工作的。如果数据集线性不可分，那么PCA不能适用
- 核主成分分析的更多介绍：https://www.ics.uci.edu/~welling/classnotes/papers_class/Kernel-PCA.pdf

```python
%run kpca.py
```

做盲源分离：
- 指将信号从混合体中分离出来的过程
- 假设一组不同的信号发生器生成了不同的信号，而一个公共接收机收到了所有这些信号。我们需要利用这些信号性质从混合体中分离出来
- 使用独立成分分析(ICA - independent components analysis)算法来实现。具体参考：http://www.mit.edu/~gari/teaching/6.555/LECTURE_NOTES/ch15_bss.pdf
